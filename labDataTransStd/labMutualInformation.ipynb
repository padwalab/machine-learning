```python
import pandas as pd
import numpy as np
import datetime
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sn
```

```python
wine = datasets.load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df.head()
```

```python
df.describe()
```

### StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
standerScaler = StandardScaler()
wineDataStandardNorm = standerScaler.fit_transform(df)
```

```python
wineDataStandardScaled = pd.DataFrame(wineDataStandardNorm, index=df.index, columns=df.columns)
wineDataStandardScaled.head()
```

```python
wineDataStandardScaled.describe()
```

```python
plt.scatter(wineDataStandardNorm[:, 0],wineDataStandardNorm[:,1])
plt.axis('equal')
```

### Standardization PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 8)
```

```python
wineDataPca = pca.fit_transform(wineDataStandardNorm)
pca.explained_variance_ratio_.sum()
```

```python
plt.scatter(wineDataPca[:,0],wineDataPca[:,1])
```

```python
wdNormStd = pd.DataFrame(pca.components_, columns=df.columns)
wdNormStd.head()
```

```python
covMatrix = np.cov(wdNormStd)
covMatrixDf = pd.DataFrame(covMatrix)
covMatrixDf
```

```python
covMatrixDf.style.background_gradient(cmap='rainbow', axis=None)
```

```python
sn.heatmap(covMatrixDf, annot=False, fmt='g')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,8))
sn.heatmap(covMatrixDf, cmap=sn.diverging_palette(220, 10, as_cmap=True), square=True,ax=ax)
```

```python
pd.plotting.scatter_matrix(covMatrixDf, alpha=0.3, figsize=(14,8),diagonal='hist')
```
